```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
dataSet = {'names': ['Steve', 'mary', 'joseph', 'Gabriel', 'Poxxie', 'Cassie'],
           'age':[20,21,19,23,24,25],
           'gender':['Male', 'Female','Male', 'Male', 'Female', 'Female'],                                         'rank':[2,1,4,3,2,1] }

df=DataFrame(dataSet)
df
```

### Matplotlib's Bar Chart

```python
plt.bar(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing ages')
```

# Bar chart seaborn

```python

plot=sns.barplot(data=df, x='names', y='age')
plot.set_title("Comparing ages")
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing ages')
plt.show()
```

### Line Plot Seaborn

```python
plot = sns.lineplot(data=df, x='names', y='age')
plot.set_title("Comparing ages")
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['age'], labels=df['names'])
plt.title('Comparing ages')
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('pastel')[0:6]
plt.pie(df['age'], labels=df['names'], colors=colors)
plt.show()
```
